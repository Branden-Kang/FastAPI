# The Real Issue

```python
@app.get("/data")
async def get_data():
    result = requests.get("https://api.example.com/data")
    return result.json()
```

# The Fix
FastAPI uses async. But requests.get() is blocking. So every single request was blocking the event loop and slowing down every other request running alongside it. Under load, everything grinds down together.

```python
import httpx

@app.get("/data")
async def get_data():
    async with httpx.AsyncClient() as client:
        result = await client.get("https://api.example.com/data")
    return result.json()
```
